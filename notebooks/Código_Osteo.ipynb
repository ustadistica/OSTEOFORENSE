### Explicación del código paso a paso:

1. **Clonación y preparación del entorno**  
   Se clona el repositorio `OSTEOFORENSE` y se cambia el directorio de trabajo a este. Luego se instalan las librerías necesarias: `dlib` para detección de landmarks faciales, `opencv-python-headless` para procesamiento de imágenes, `matplotlib` para mostrar imágenes, `numpy` para cálculos matriciales, y `openpyxl` para trabajar con archivos Excel.

2. **Descarga del modelo predictivo de landmarks**  
   Se descarga y descomprime el archivo `shape_predictor_68_face_landmarks.dat` que contiene el modelo pre-entrenado para detectar 68 puntos faciales clave (landmarks).

3. **Funciones de carga y procesamiento de imágenes**  
   - `cargar_imagenes`: carga todas las imágenes con extensión jpg/png del directorio, devolviendo una lista de imágenes y sus nombres.  
   - `obtener_landmarks`: para cada imagen, convierte a escala de grises, detecta la cara y predice los 68 puntos faciales según el modelo.  
   - `distancia`: calcula la distancia euclidiana entre dos puntos (útil para medir entre landmarks).  
   - `dibujar_medidas`: dibuja líneas rojas entre pares de landmarks que definen las medidas a obtener y puntos amarillos para visualización.

4. **Diccionario de pares de landmarks para calcular medidas faciales**  
   Se define un conjunto de medidas relevantes para antropometría forense como altura nasal, ancho nasal, ancho boca, distancia bizigomática, bigonial, etc., usando índices de puntos en el array de landmarks.

5. **Función principal `procesar_y_medir`**  
   Esta función:  
   - Inicializa los detectores y predictores de dlib con el modelo.  
   - Carga imágenes del directorio.  
   - Crea un archivo Excel nuevo para registrar resultados.  
   - Para cada imagen:  
     - Obtiene landmarks.  
     - Si no se detecta rostro, se omite la imagen.  
     - Calcula una escala usando la medida bi-ectocanthal (ex-ex = 96 mm).  
     - Calcula las distancias en pixeles y las convierte a milímetros.  
     - Imprime las medidas por imagen.  
     - Dibuja las líneas y puntos sobre la imagen para mostrar visualmente las medidas.  
     - Guarda temporalmente la imagen procesada para luego insertarla en Excel.  
     - Añade al Excel una fila con nombre imagen, medidas en px, medidas en mm y la imagen procesada incrustada.  
   - Finalmente, añade una hoja resumen con la cantidad total de imágenes procesadas y el modelo usado.  
   - Guarda el archivo Excel resultante con todos los datos y las imágenes anotadas.

En resumen, el código automatiza el proceso de análisis morfométrico facial osteoforense: detecta puntos anatómicos, mide distancias clave, convierte a unidades reales con una referencia, visualiza resultados y los registra integralmente en un archivo Excel para trazabilidad y análisis posterior.


```python
# Clonar el repositorio OSTEOFORENSE
!rm -rf OSTEOFORENSE # borra si existe para evitar error
!git clone https://github.com/ustadistica/OSTEOFORENSE.git

# Cambiar al directorio del repo
%cd OSTEOFORENSE

# Instalar dependencias necesarias
!pip install dlib opencv-python-headless matplotlib numpy openpyxl --quiet

# Descargar modelo landmarks dlib
!wget http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2
!bzip2 -d shape_predictor_68_face_landmarks.dat.bz2

# Código Python para análisis osteo-facial con medidas, visualización y guardado en Excel

import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt
import os
import tempfile
from openpyxl import Workbook
from openpyxl.drawing.image import Image as ExcelImage

# Ruta imágenes dentro del repo
carpeta_imagenes = "data/imagenes"
predictor_path = "shape_predictor_68_face_landmarks.dat"

def cargar_imagenes(directorio):
    imagenes, nombres = [], []
    for fname in os.listdir(directorio):
        if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
            path = os.path.join(directorio, fname)
            img = cv2.imread(path)
            if img is not None:
                imagenes.append(img)
                nombres.append(os.path.splitext(fname)[0])
    print(f"{len(imagenes)} imágenes cargadas desde {directorio}")
    return imagenes, nombres

def obtener_landmarks(imagen, detector, predictor):
    gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    rects = detector(gray, 1)
    if len(rects) == 0:
        return None
    shape = predictor(gray, rects[0])
    coords = np.array([(shape.part(i).x, shape.part(i).y) for i in range(68)])
    return coords

def distancia(p1, p2):
    return np.linalg.norm(p1 - p2)

def dibujar_medidas(img, landmarks, medida_puntos, color_rojo=(0, 0, 255), color_amarillo=(0, 255, 255)):
    img_out = img.copy()
    # Líneas rojas para medidas
    for (p1, p2) in medida_puntos.values():
        cv2.line(img_out, tuple(landmarks[p1]), tuple(landmarks[p2]), color_rojo, 1)
    # Puntos amarillos (radio pequeño)
    for i, (x, y) in enumerate(landmarks):
        cv2.circle(img_out, (x, y), 2, color_amarillo, -1)
    return img_out

medida_puntos = {
    'n_sn': (27, 33),    # altura nasal
    'al_al': (31, 35),   # ancho nasal
    'en_en': (39, 42),   # intercanthal
    'ex_ex': (36, 45),   # bi-ectocantal
    'ch_ch': (48, 54),   # ancho boca
    'zy_zy': (1, 15),    # bizigomático
    'go_go': (5, 11),    # bigonial aprox
    'g_gn': (27, 8),     # altura facial
    'sn_gn': (33, 8),    # tercio inferior
}

def procesar_y_medir(directorio, predictor_path, escala_exex_mm=96):
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(predictor_path)
    imagenes, nombres = cargar_imagenes(directorio)

    # Crear libro de Excel para guardar resultados
    wb = Workbook()
    ws = wb.active
    ws.title = "Medidas Osteofacial"

    # Encabezados
    headers = ["Nombre Imagen"] + [f"{m} px" for m in medida_puntos.keys()] + [f"{m} mm" for m in medida_puntos.keys()] + ["Imagen Procesada"]
    ws.append(headers)

    total_imagenes_procesadas = 0

    for img, nombre in zip(imagenes, nombres):
        landmarks = obtener_landmarks(img, detector, predictor)
        if landmarks is None:
            print(f"No se detectó rostro en {nombre}")
            continue

        pixeles, mm = {}, {}

        base_px = distancia(landmarks[36], landmarks[45])
        escala = escala_exex_mm / base_px

        for medida, (p1, p2) in medida_puntos.items():
            d = distancia(landmarks[p1], landmarks[p2])
            pixeles[medida] = d
            mm[medida] = d * escala

        print(f"Medidas para imagen {nombre} (ex ex={escala_exex_mm} mm):")
        for m in medida_puntos.keys():
            print(f" {m}  px: {pixeles[m]:.1f}  mm: {mm[m]:.1f}")

        img_medidas = dibujar_medidas(img, landmarks, medida_puntos)

        # Mostrar imagen procesada
        plt.figure(figsize=(6, 6))
        plt.title(nombre, fontsize=8)
        plt.imshow(cv2.cvtColor(img_medidas, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

        # Guardar imagen temporal para insertar en Excel
        tmp_file = tempfile.NamedTemporaryFile(suffix=".png", delete=False)
        cv2.imwrite(tmp_file.name, img_medidas)
        tmp_file.close()

        # Preparar fila con datos de medidas para Excel
        fila = [nombre]
        fila += [round(pixeles[m], 2) for m in medida_puntos.keys()]
        fila += [round(mm[m], 2) for m in medida_puntos.keys()]
        ws.append(fila)

        # Insertar imagen en la última fila (columna correspondiente a "Imagen Procesada")
        img_excel = ExcelImage(tmp_file.name)
        img_excel.width = 160
        img_excel.height = 160

        col_img = chr(65 + len(headers) - 1)  # Columna para la imagen (letra)
        ws.add_image(img_excel, f"{col_img}{ws.max_row}")

        total_imagenes_procesadas += 1

    # Añadir una hoja resumen con total de imágenes y modelo usado
    resumen = wb.create_sheet(title="Resumen")
    resumen.append(["Total imágenes procesadas", total_imagenes_procesadas])
    resumen.append(["Modelo de landmarks usado", predictor_path])

    # Guardar archivo Excel
    nombre_excel = "Resultados_OsteoFacial.xlsx"
    wb.save(nombre_excel)
    print(f"Archivo Excel guardado: {nombre_excel}")

# Ejecutar análisis completo
procesar_y_medir(carpeta_imagenes, predictor_path)
```
